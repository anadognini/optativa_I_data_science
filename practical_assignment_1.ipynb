## 1. Short Report

### Dataset motivation and description

The primary motivation for utilizing this dataset is to explore the intricate relationship between daily lifestyle habits and sleep quality. Sleep is a fundamental pillar of biological health, yet it is often compromised by modern stressors and occupational demands. This dataset provides a comprehensive synthetic profile of 374 individuals, mapping variables such as physical activity, stress levels, and BMI against sleep metrics like duration and quality. By analyzing these patterns, we aim to identify the key predictors of sleep disorders and understand how lifestyle choices influence overall well-being.

### Data Dictionary Explanation

The dataset consists of 13 columns covering demographic, lifestyle, and cardiovascular health information:

- Demographics: Includes Person ID, Gender, and Age.

- Lifestyle Metrics: Occupation, Sleep Duration (hours), Quality of Sleep (1-10 scale), Physical Activity Level (minutes/day), and Daily Steps.

- Health & Stress: Stress Level (1-10 scale), BMI Category, Blood Pressure (systolic/diastolic), and Heart Rate (bpm).

- Target Variable: Sleep Disorder, which categorizes individuals into "None", "Insomnia", or "Sleep Apnea".

### Cleaning Procedures and Decisions

The data cleaning process was straightforward due to the high initial quality of the source file. The primary technical intervention involved handling missing values within the Sleep Disorder column.

In the original dataset, individuals without a diagnosed sleep disorder were represented by null values (NaN). To ensure compatibility with machine learning algorithms and improve readability during exploratory data analysis (EDA), these were systematically replaced with the string "None". This maintains the categorical integrity of the column without losing data points.

### Initial Data Quality Discussion

The dataset exhibits high data quality and structural integrity. Upon initial inspection, no significant anomalies, such as impossible outliers (e.g., negative ages or sleep hours) or inconsistent formatting, were found. The distributions for numerical variables like Heart Rate and Stress Level are within expected physiological ranges. Since the data is already well-curated, no intensive noise reduction or row-level deletion was required, allowing the focus to remain on feature engineering and statistical modeling.
